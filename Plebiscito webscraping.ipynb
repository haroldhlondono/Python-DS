```python
from bs4 import BeautifulSoup
```

```python
import requests
```

Esta función consigue html del website

```python
def get_soup(s):
    r = requests.get(s)
    data = r.text
    soup = BeautifulSoup(data, "lxml")
    return soup

soup = get_soup(url)
```

## Obtener información de los departamentos
### Cada elemento de la lista es tipo bs4.element.Tag

```python
departments_soup = soup.find_all("option")[56:90]
```

```python
import pandas as pd
```

```python
departments = pd.DataFrame(columns=("Department", "path"))

for i in range (0,len(departments_soup)):
    row=pd.Series(dict(zip(["Department", "path"],
                           [departments_soup[i].get_text(),
                            departments_soup[i]["value"]])))
    row.name = i
    departments = departments.append(row)

```
