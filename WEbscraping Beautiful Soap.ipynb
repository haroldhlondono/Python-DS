## Extracción con Beautiful Soup

```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/74.0.3729.157 Safari/537.36"
}
```

```python
url = "https://stackoverflow.com/questions"
```

```python
respuesta = requests.get(url, headers = headers)
```

```python
soup = BeautifulSoup(respuesta.text)


```

```python
contenedor_de_preguntas = soup.find(id="questions")
lista_de_preguntas = contenedor_de_preguntas.find_all("div", class_="question-summary")

for pregunta in lista_de_preguntas:
    texto_pregunta = pregunta.find("h3").text
    print(texto_pregunta)
```

```python
contenedor_de_preguntas = soup.find(id="questions")
lista_de_preguntas = contenedor_de_preguntas.find_all("div", class_="question-summary")

for pregunta in lista_de_preguntas:
    texto_pregunta = pregunta.find("h3").text
    descripcion_pregunta = pregunta.find(class_="excerpt").text
    print(texto_pregunta)
    print(descripcion_pregunta)
    
    print()
```

Comando para limpiar un pcoco los datos

```python
contenedor_de_preguntas = soup.find(id="questions")
lista_de_preguntas = contenedor_de_preguntas.find_all("div", class_="question-summary")

for pregunta in lista_de_preguntas:
    texto_pregunta = pregunta.find("h3").text
    descripcion_pregunta = pregunta.find(class_="excerpt").text
    descripcion_pregunta = descripcion_pregunta.replace("\n", '').replace('\r', '').strip() ## esta linea
    print(texto_pregunta)
    print(descripcion_pregunta)
    
    print()
```

### Beautiful Soup es muy util para decirle que lo mande al primo más cercano, o padre o hijo

```python
contenedor_de_preguntas = soup.find(id="questions")
lista_de_preguntas = contenedor_de_preguntas.find_all("div", class_="question-summary")

for pregunta in lista_de_preguntas:
    elemento_texto_pregunta = pregunta.find("h3") ##esta linea
    texto_pregunta = elemento_texto_pregunta.text
    
    elemento_texto_pregunta.find_next_sibling("div").text
    
    descripcion_pregunta = pregunta.find(class_="excerpt").text
    descripcion_pregunta = descripcion_pregunta.replace("\n", '').replace('\r', '').strip() ## esta linea
    print(texto_pregunta)
    print(descripcion_pregunta)
    
    print()
```
